# split, training, predict and mean square error

```python
data_ordinal = pd.read_csv('data_transformation_dataset_task2.csv', index_col=0)
data_ordinal
```

```python
def ordinal(x):
    if x == 'A':
        return 1
    elif x == 'B':
        return 2
    elif x == 'C':
        return 3
    elif x=='D':
        return 4
    elif x == 'E':
        return 5
    elif x == 'F':
        return 6
    elif x=='G':
        return 7
    elif x == 'H':
        return 8
    elif x == 'I':
        return 9
    elif x=='J':
        return 10
    elif x == 'K':
        return 11
    elif x == 'L':
        return 12
    elif x=='M':
        return 13
    elif x == 'N':
        return 14
    elif x=='O':
        return 15
    elif x == 'P':
        return 16
   
  
    else:
        return np.NaN

data_ordinal['cat0'] = data_ordinal['cat0'].apply(ordinal)
data_ordinal['cat1'] = data_ordinal['cat1'].apply(ordinal)
data_ordinal['cat2'] = data_ordinal['cat2'].apply(ordinal)
data_ordinal['cat3'] = data_ordinal['cat3'].apply(ordinal)
data_ordinal['cat4'] = data_ordinal['cat4'].apply(ordinal)
data_ordinal['cat5'] = data_ordinal['cat5'].apply(ordinal)
data_ordinal['cat6'] = data_ordinal['cat6'].apply(ordinal)
data_ordinal['cat7'] = data_ordinal['cat7'].apply(ordinal)
data_ordinal['cat8'] = data_ordinal['cat8'].apply(ordinal)
data_ordinal['cat9'] = data_ordinal['cat9'].apply(ordinal)
```

```python
data_ordinal.head(5)
```

```python
data_ordinal = data_ordinal.astype(float)
#data_ordinal = pd.to_numeric(data_ordinal)
```

```python
y = data_ordinal['target']# what you want predict 
X = data_ordinal.drop(['target'], axis=1) # predictors
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

X_train1, X_test1, y_train1, y_test1 = train_test_split(X, y,  test_size=0.33, random_state=42)

model = RandomForestRegressor(n_estimators = 100,
                           n_jobs = -1,
                           oob_score = True,
                           bootstrap = True,
                           random_state = 42) 
model.fit(X_train1, y_train1)    
y_pred = model.predict(X_test1)

r2_score(y_test1,y_pred)
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test1,y_pred, squared=False))
```
